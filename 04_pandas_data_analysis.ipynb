## Part 1: Loading Data

Pandas makes it easy to load data from various formats. Let's start with CSV files.

```python
import pandas as pd
import numpy as np

# Create sample data for demonstration
sample_data = {
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'Philadelphia', 'San Antonio', 'San Diego'],
    'State': ['NY', 'CA', 'IL', 'TX', 'AZ', 'PA', 'TX', 'CA'],
    'Population': [8336817, 3979576, 2693976, 2320268, 1680992, 1584064, 1547253, 1423851],
    'Latitude': [40.7128, 34.0522, 41.8781, 29.7604, 33.4484, 39.9526, 29.4241, 32.7157],
    'Longitude': [-74.0060, -118.2437, -87.6298, -95.3698, -112.0740, -75.1652, -98.4936, -117.1611],
    'Median_Income': [63998, 62474, 58247, 52338, 59206, 45927, 52455, 79673]
}

df = pd.DataFrame(sample_data)

# Save to CSV for loading demonstration
df.to_csv('cities_data.csv', index=False)
print("Sample data created and saved to cities_data.csv")
```

```python
# Load the CSV file
cities_df = pd.read_csv('cities_data.csv')

print("Data loaded successfully!")
print(f"\nDataFrame shape: {cities_df.shape}")
print(f"Rows: {cities_df.shape[0]}, Columns: {cities_df.shape[1]}")
```

## Part 3: Data Analysis

Let's perform some basic analysis on our dataset.

```python
# Calculate basic statistics
print("=== Population Analysis ===")
print(f"Average population: {cities_df['Population'].mean():,.0f}")
print(f"Median population: {cities_df['Population'].median():,.0f}")
print(f"Total population: {cities_df['Population'].sum():,.0f}")
print(f"Largest city: {cities_df.loc[cities_df['Population'].idxmax(), 'City']}")
print(f"Smallest city: {cities_df.loc[cities_df['Population'].idxmin(), 'City']}")
```

```python
# Group by state
print("\n=== Cities by State ===")
state_counts = cities_df['State'].value_counts()
print(state_counts)

print("\n=== Average Population by State ===")
state_pop = cities_df.groupby('State')['Population'].agg(['mean', 'sum'])
state_pop.columns = ['Average Population', 'Total Population']
display(state_pop)
```

```python
# Filter data
print("=== Cities with Population > 2 Million ===")
large_cities = cities_df[cities_df['Population'] > 2000000]
display(large_cities[['City', 'State', 'Population']])
```

```python
# Sort data
print("=== Top 5 Cities by Median Income ===")
top_income = cities_df.nlargest(5, 'Median_Income')[['City', 'State', 'Median_Income']]
display(top_income)
```

## Part 4: Data Visualization with Pandas

Pandas integrates with Matplotlib for quick visualizations.

```python
import matplotlib.pyplot as plt

# Set style for better-looking plots
plt.style.use('seaborn-v0_8-darkgrid')

# Bar chart of populations
cities_df.plot(x='City', y='Population', kind='bar', figsize=(12, 6), color='skyblue')
plt.title('City Populations', fontsize=16, fontweight='bold')
plt.xlabel('City', fontsize=12)
plt.ylabel('Population', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Scatter plot: Population vs Median Income
plt.figure(figsize=(10, 6))
plt.scatter(cities_df['Population'], cities_df['Median_Income'], 
            s=200, alpha=0.6, c='coral', edgecolors='black', linewidth=1.5)

# Add city labels
for idx, row in cities_df.iterrows():
    plt.annotate(row['City'], 
                (row['Population'], row['Median_Income']),
                xytext=(5, 5), textcoords='offset points', fontsize=9)

plt.title('Population vs Median Income', fontsize=16, fontweight='bold')
plt.xlabel('Population', fontsize=12)
plt.ylabel('Median Income ($)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Pie chart of cities by state
state_counts = cities_df['State'].value_counts()
plt.figure(figsize=(10, 8))
plt.pie(state_counts.values, labels=state_counts.index, autopct='%1.1f%%',
        startangle=90, colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'])
plt.title('Distribution of Cities by State', fontsize=16, fontweight='bold')
plt.axis('equal')
plt.show()
```

## Part 5: Combining Pandas and Folium

Now for the exciting part - let's put our data on a map!

```python
import folium

# Create a map centered on the US
cities_map = folium.Map(
    location=[37.0902, -95.7129],  # Center of US
    zoom_start=4,
    tiles='CartoDB positron'
)

# Add markers for each city
for idx, row in cities_df.iterrows():
    # Create popup HTML
    popup_html = f"""
    <div style="font-family: Arial; font-size: 12px; width: 200px;">
        <h4 style="margin-bottom: 5px;">{row['City']}, {row['State']}</h4>
        <p style="margin: 3px 0;"><strong>Population:</strong> {row['Population']:,}</p>
        <p style="margin: 3px 0;"><strong>Median Income:</strong> ${row['Median_Income']:,}</p>
        <p style="margin: 3px 0;"><strong>Coordinates:</strong> {row['Latitude']:.4f}, {row['Longitude']:.4f}</p>
    </div>
    """
    
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=folium.Popup(popup_html, max_width=250),
        tooltip=row['City'],
        icon=folium.Icon(color='blue', icon='info-sign')
    ).add_to(cities_map)

cities_map
```

```python
# Create a map with circle markers sized by population
population_map = folium.Map(
    location=[37.0902, -95.7129],
    zoom_start=4,
    tiles='OpenStreetMap'
)

# Normalize population for circle size
max_pop = cities_df['Population'].max()
min_pop = cities_df['Population'].min()

for idx, row in cities_df.iterrows():
    # Scale population to circle radius (10-50 pixels)
    radius = 10 + (row['Population'] - min_pop) / (max_pop - min_pop) * 40
    
    # Color based on income
    if row['Median_Income'] > 60000:
        color = 'green'
    elif row['Median_Income'] > 50000:
        color = 'orange'
    else:
        color = 'red'
    
    popup_html = f"""
    <strong>{row['City']}</strong><br>
    Pop: {row['Population']:,}<br>
    Income: ${row['Median_Income']:,}
    """
    
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=radius,
        popup=popup_html,
        tooltip=row['City'],
        color=color,
        fill=True,
        fillColor=color,
        fillOpacity=0.7
    ).add_to(population_map)

# Add a legend
legend_html = '''
<div style="position: fixed; 
     bottom: 50px; left: 50px; width: 180px; height: 120px; 
     background-color: white; border:2px solid grey; z-index:9999; 
     font-size:14px; padding: 10px">
     <p style="margin: 5px;"><strong>Legend</strong></p>
     <p style="margin: 5px;">ðŸŸ¢ Income > $60k</p>
     <p style="margin: 5px;">ðŸŸ  Income $50k-$60k</p>
     <p style="margin: 5px;">ðŸ”´ Income < $50k</p>
     <p style="margin: 5px;">Size = Population</p>
</div>
'''
population_map.get_root().html.add_child(folium.Element(legend_html))

population_map
```

## Part 6: Advanced Analysis - Correlation

Let's explore relationships between variables.

```python
# Calculate correlation matrix
numeric_cols = ['Population', 'Median_Income']
correlation = cities_df[numeric_cols].corr()

print("Correlation Matrix:")
display(correlation)

# Visualize correlation
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0, 
            square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Correlation Heatmap', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```
